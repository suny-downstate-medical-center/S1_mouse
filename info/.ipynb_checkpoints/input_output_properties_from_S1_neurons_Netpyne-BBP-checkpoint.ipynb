```python
os.chdir(rootFolder)
import neuron as nrn # NEURON simulator
nrn.h.load_file("./init.hoc");
```

```python
cell = nrn.h.cell
soma = cell.soma[0]
```

```python
stimulus = nrn.h.IClamp(0.5, sec=soma)
```

```python
stimulus.dur = 1000 # ms
stimulus.delay = 100  # ms     
```

```python
with open('current_amps.dat') as current_file:
    current_content = current_file.read()

print 'File content: ', current_content
holding_current, step1_current, step2_current, step3_current = [float(x) for x in current_content.split()]

print
print 'Holding current: %f nA' % holding_current
print 'Step 1: %f nA' % step1_current
print 'Step 2: %f nA' % step2_current
print 'Step 3: %f nA' % step3_current
```

```python
# fig = plt.figure(figsize=(9,9))
# fig.subplots_adjust(hspace=0.4, wspace=0.4)

bmtkTraces = []
bmtkTracesList = []

i=0
for x in current_content.split():
    i=i+1   
    nrn.h.create_recording();    #activate recording of activity 
    
    if float(x)<0:
        stimulus.amp = float(x)
    else:
        stimulus.amp = holding_current+float(x)
        
    nrn.h.dt = 0.05
    
    nrn.h.tstop = 1200 # ms
    nrn.h.run();
    nrn.h.save_recording()
    time = nrn.h.time
    voltage = nrn.h.voltage
    bmtkTraces.append(np.array(voltage))
    bmtkTracesList.append(list(voltage))
#     ax = fig.add_subplot(4, 1, i)
#     ax.plot(time,voltage)
#     plt.xlabel('Time (ms)')
#     plt.ylabel('V_soma (mV)')
#     plt.xlim(0,1200)

# fig.savefig('{s1}/{s}.png'.format(s1=rootFolder,s=folder[cellnumber]))
```

```python
if runPlot:
    from netpyne import sim
    from netpyne import specs
    import pickle

    cfg = specs.SimConfig()  

    #------------------------------------------------------------------------------
    #
    # SIMULATION CONFIGURATION
    #
    #------------------------------------------------------------------------------

    #------------------------------------------------------------------------------
    # Run parameters
    #------------------------------------------------------------------------------

    cfg.duration = 1.2*1e3			## Duration of the sim, in ms -- value from M1 cfg.py 
    cfg.dt = 0.05                   ## Internal Integration Time Step -- value from M1 cfg.py 
    cfg.verbose = False           	## Show detailed messages
    cfg.seeds = {'conn': 4321, 'stim': 1234, 'loc': 4321} 


    #------------------------------------------------------------------------------
    # Recording 
    #------------------------------------------------------------------------------

    allpops = ['L1_1','L1_2','L1_3','L1_4']

    cfg.recordCells = allpops  # which cells to record from
    cfg.recordTraces = {'V_soma': {'sec':'soma_0', 'loc':0.5, 'var':'v'}}  ## Dict with traces to record -- taken from M1 cfg.py 
    cfg.recordStim = True			## Seen in M1 cfg.py
    cfg.recordTime = True  		## SEen in M1 cfg.py 
    cfg.recordStep = 0.1            ## Step size (in ms) to save data -- value from M1 cfg.py 



    #------------------------------------------------------------------------------
    # Saving
    #------------------------------------------------------------------------------

    cfg.simLabel = 'S1detailed'
    cfg.saveFolder = '.'
    # cfg.filename =                	## Set file output name
    cfg.savePickle = True         	## Save pkl file
    cfg.saveJson = True           	## Save json file
    cfg.saveDataInclude = ['simConfig', 'netParams'] ## seen in M1 cfg.py (line 58) 'simData' , 'simConfig', 'netParams'
    cfg.backupCfgFile = None 		## Seen in M1 cfg.py 
    cfg.gatherOnlySimData = False	## Seen in M1 cfg.py 
    cfg.saveCellSecs = True			## Seen in M1 cfg.py 
    cfg.saveCellConns = True		## Seen in M1 cfg.py 

    #------------------------------------------------------------------------------
    # Analysis and plotting 
    #------------------------------------------------------------------------------
    #cfg.analysis['plotTraces'] = {'include': [('L1_1',0), ('L1_2',0),('L1_3',0), ('L1_4',0)], 'saveFig': True, 'showFig': False, 'oneFigPer':'trace', 'overlay':False} 		

    #cfg.analysis['plotShape'] = False

    #------------------------------------------------------------------------------
    # Cells
    #------------------------------------------------------------------------------
    cfg.cellmod =  {'L1_1': 'HH_full'}
    cfg.cellmod =  {'L1_2': 'HH_full'}
    cfg.cellmod =  {'L1_3': 'HH_full'}
    cfg.cellmod =  {'L1_4': 'HH_full'}


    #------------------------------------------------------------------------------
    # Synapses
    #------------------------------------------------------------------------------


    cfg.AMPATau2Factor = 1.0


    #------------------------------------------------------------------------------
    # Subcellular distribution
    #------------------------------------------------------------------------------
    cfg.addSubConn = 1

    #------------------------------------------------------------------------------
    # Current inputs 
    #------------------------------------------------------------------------------
    cfg.addIClamp = 1

    cfg.IClamp1 = {'pop': 'L1_1', 'sec': 'soma_0', 'loc': 0.5, 'start': 100, 'dur': 1000, 'amp': holding_current}
    cfg.IClamp2 = {'pop': 'L1_2', 'sec': 'soma_0', 'loc': 0.5, 'start': 100, 'dur': 1000, 'amp': holding_current+step1_current}
    cfg.IClamp3 = {'pop': 'L1_3', 'sec': 'soma_0', 'loc': 0.5, 'start': 100, 'dur': 1000, 'amp': holding_current+step2_current}
    cfg.IClamp4 = {'pop': 'L1_4', 'sec': 'soma_0', 'loc': 0.5, 'start': 100, 'dur': 1000, 'amp': holding_current+step3_current}


    #------------------------------------------------------------------------------
    # NetStim inputs 
    #------------------------------------------------------------------------------

    ## Attempt to add Background Noise inputs 
    cfg.addNetStim = 0


    """
    netParams.py

    High-level specifications for M1 network model using NetPyNE

    Contributors: salvadordura@gmail.com
    """

    from netpyne import specs
    import pickle, json
    from neuron import gui
    import matplotlib.pyplot as plt
    import numpy as np


    netParams = specs.NetParams()   # object of class NetParams to store the network parameters


    try:
        from __main__ import cfg  # import SimConfig object with params from parent module
    except:
        from cfg import cfg


    #------------------------------------------------------------------------------
    # Cell parameters
    #------------------------------------------------------------------------------

    cellRule = netParams.importCellParams(label=cellName + '_rule',
        conds={'cellType': cellName, 'cellModel': 'HH_full'},
        fileName='cellwrapper.py',
        cellName='loadCell',
        cellInstance = True,
        cellArgs={'cellName': cellName, 'cellTemplateName': cellTemplateName})


    #------------------------------------------------------------------------------
    # Population parameters
    #------------------------------------------------------------------------------

    netParams.popParams['L1_1'] = {'cellType': cellName, 'cellModel': 'HH_full', 'numCells': 1} 
    netParams.popParams['L1_2'] = {'cellType': cellName, 'cellModel': 'HH_full', 'numCells': 1} 
    netParams.popParams['L1_3'] = {'cellType': cellName, 'cellModel': 'HH_full', 'numCells': 1} 
    netParams.popParams['L1_4'] = {'cellType': cellName, 'cellModel': 'HH_full', 'numCells': 1} 


    #------------------------------------------------------------------------------
    # Current inputs (IClamp)
    #------------------------------------------------------------------------------
    if cfg.addIClamp:
         for key in [k for k in dir(cfg) if k.startswith('IClamp')]:
            params = getattr(cfg, key, None)
            [pop,sec,loc,start,dur,amp] = [params[s] for s in ['pop','sec','loc','start','dur','amp']]

            #cfg.analysis['plotTraces']['include'].append((pop,0))  # record that pop

            # add stim source
            netParams.stimSourceParams[key] = {'type': 'IClamp', 'delay': start, 'dur': dur, 'amp': amp}

            # connect stim source to target
            netParams.stimTargetParams[key+'_'+pop] =  {
                'source': key, 
                'conds': {'pop': pop},
                'sec': sec, 
                'loc': loc}
    

    sim.createSimulateAnalyze(netParams, cfg)
```

```python
if compareTraces:
    # store netpyne traces
    netpyneTraces = []
    netpyneTracesList = []
    for c in range(0,4):
        netpyneTraces.append(np.array(sim.simData['V_soma']['cell_'+ str(c)]))
        netpyneTracesList.append(list(sim.simData['V_soma']['cell_'+ str(c)]))

    with open(outFolder+'/netpyne_traces_300cells.json', 'w') as f:
        json.dump(netpyneTracesList, f)

    # plot both traces overlayed
    recordStep = sim.cfg.recordStep
    timeRange = [0, sim.cfg.duration]
    fontsiz=8
    ylim = [-100, 40]
    figSize = (10,10)
    fig = plt.figure(figsize=figSize)  # Open a new figure

    for c in range(0,4):
        netpyneTrace = netpyneTraces[c]
        bmtkTrace = bmtkTraces[c]
        t = np.arange(timeRange[0], timeRange[1]+recordStep, recordStep)
        plt.subplot(len(sim.cfg.recordCells), 1, c+1)
        plt.ylabel('V (mV)', fontsize=fontsiz)
        plt.plot(t[:len(netpyneTrace)], netpyneTrace, linewidth=1.5, color='red', label='Step %d'%(int(c))+', NetPyNE')
        plt.plot(t[:len(bmtkTrace)], bmtkTrace, linewidth=1.0, color='green', label='Step %d'%(int(c))+', BBP')  # linestyle=':'
        plt.xlabel('Time (ms)', fontsize=fontsiz)
        plt.xlim(timeRange)
        plt.ylim(ylim)
        plt.grid(True)
        plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1.0))
    plt.ion()
    plt.tight_layout()
    plt.savefig(outFolder+'/comparison_traces.png')
    plt.show()
```
