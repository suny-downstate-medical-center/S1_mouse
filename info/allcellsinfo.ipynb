```python
def volume(sizey, radius):    
    sizey = 0.001 * sizey # from um to mm
    radius = 0.001 * radius # from um to mm    
    vol = np.pi * radius**2 * sizey
    return vol

def cellNumber(sizey, radius, density):
    number = volume(sizey, radius) * density
    return number
```
